```python
# creating a function for visualizing the credit amount w.r.t target
def grouping_col(df, col, target):
    """The function uses col for grouping dataframe and shows the visualization - distribution of target variable and count plots"""
    df_temp = df.loc[:,[col, target]]
    if (is_numeric_dtype(df[col])) and (df[col].nunique() >10):
        if df_temp[col].isnull().sum() > 0:
            df_temp = df_temp[~df_temp[col].isnull()]
        fig, (ax1, ax2) = plt.subplots(1,2, figsize = (16,6))
        sns.regplot(y = df_temp[target], x = df_temp[col], ax = ax1)
        sns.distplot(df_temp[col], kde = False)
        plt.show()
    
    else:
        df_temp[col] = df_temp[col].fillna('NA')
#         df_temp = df_temp.groupby(col)[target].median().reset_index()
        fig, (ax1, ax2) = plt.subplots(1,2, figsize = (16,6))
        sns.boxplot(y = df_temp[target], x = df_temp[col], ax = ax1)
        sns.countplot(x = df_temp[col], ax = ax2)
        plt.show()
```

```python
# creating functions which could be used for crossvalidation
# also, created a metric score as indicated in the problem statement
def metric_results(model, X_train, y_train):
    model.fit(X_train,y_train)
    cv_score = cross_val_score(model, X_train, y_train, cv = folds, n_jobs = -1, scoring = metric_scorer)
    return cv_score, cv_score.mean(), cv_score.std()

def metric(y_true, y_pred):
#     y_true = np.exp(y_true)
#     y_pred = np.exp(y_pred)
    mse = mean_squared_error(y_true, y_pred)
    return (1- np.sqrt(mse)/100000)*100

metric_scorer = make_scorer(metric)
```

```python
# defining xgb_fit for utilizing cv function in the xgb module and use the cross validation for obtaining optimum n_estimators

kfolds=KFold(n_splits=5, shuffle=True, random_state=16).get_n_splits(X_train)
def xgb_fit(xgb_m, X_train, y_train, req_cv = True, esr = 50 ):
    if req_cv:
        params = xgb_m.get_xgb_params()
        xg_train = xgb.DMatrix(X_train.values, label = y_train)
        cv_result = xgb.cv(params, xg_train, num_boost_round = xgb_m.get_params()['n_estimators'], 
                          nfold = kfolds, metrics = 'rmse', early_stopping_rounds = esr, verbose_eval = True, 
                           as_pandas = True, seed = 0)
        xgb_m.set_params(n_estimators = cv_result.shape[0])
    
    xgb_m.fit(X_train, y_train, eval_metric = 'rmse')
    
#     classification_report(y_train, X_train, xgb_m)
    
#     feat_imp = pd.DataFrame({'features': X_train.columns, 'imp':xgb_m.feature_importances_})
#     feat_imp.plot(kind = 'bar')
#     plt.show()
#     print(cv_result.mean())
    return cv_result
```
